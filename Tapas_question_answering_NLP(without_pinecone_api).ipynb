```python
from transformers import pipeline
import pandas as pd
```

```python
tqa = pipeline(task="table-question-answering",
               model="google/tapas-base-finetuned-wtq")
```

```python
table = pd.read_excel("/content/CRTD_BUDGET_AOP.xlsx")
table = table.astype(str)
```

```python
table.head(50)
```

```python
query = "What is the total budget in Canada through e-comm?"
print(tqa(table=table, query=query)["answer"])
```

```python
query = "What is the total budget in Mexico through wholesaler in body care domain?"
print(tqa(table=table, query=query)["answer"])
```

```python
raw_answer = tqa(table=table, query=query)["answer"]
if raw_answer.replace(".", "", 1).isdigit():
    # If the answer is numeric, format it
    formatted_answer = f"${float(raw_answer):,.2f}"
else:
    formatted_answer = raw_answer
print(f"Answer: {formatted_answer}")
```
